```python
import cv2
import os

base_image_dir = "/content/drive/MyDrive/turn_tune/upright_images/images"
base_label_dir = "/content/drive/MyDrive/turn_tune/upright_images/labels"

splits = ["train", "val"]

for split in splits:
    image_dir = os.path.join(base_image_dir, split)
    label_dir = os.path.join(base_label_dir, split)

    os.makedirs(label_dir, exist_ok=True)

    if not os.path.exists(image_dir):
        print(f"Skipping {split}: directory not found.")
        continue

    print(f"Processing {split} set...")

    for img_name in os.listdir(image_dir):
        if not img_name.lower().endswith((".jpg", ".png", ".jpeg")):
            continue

        img_path = os.path.join(image_dir, img_name)
        img = cv2.imread(img_path)

        if img is None:
            print(f"Warning: Could not read {img_name}. Skipping.")
            continue

        h, w = img.shape[:2]

        cx, cy = 0.5, 0.5
        bw, bh = 1.0, 1.0
        x12, y12 = 0.5, 0.0

        label = f"0 {cx} {cy} {bw} {bh} {x12} {y12} 2\n"

        file_basename = os.path.splitext(img_name)[0]
        label_path = os.path.join(label_dir, f"{file_basename}.txt")

        with open(label_path, "w") as f:
            f.write(label)

print("Done! Labels generated for train and val splits.")
```

```python
model = YOLO("yolo11n-pose.pt")
```

```python
results = model.train(data = r"/content/drive/MyDrive/turn_tune/upright_images/data.yaml", epochs = 100, degrees = 180, patience = 10)
```
