```python
model_before_tuned = YOLO("yolo11n.pt")
model_after_tuned = YOLO(r"/content/drive/MyDrive/runs/detect/train13/weights/best.pt")
```

```python
metrics_before_tuned = model_before_tuned.val(
    data=r"/content/drive/MyDrive/object_detection_tuning/Clock_3/Clock_3_YOLO_original/data_yolo_original.yaml"
)
print (metrics_before_tuned.box.map75)
metrics_after_tuned = model_after_tuned.val(
    data="/content/drive/MyDrive/object_detection_tuning/Clock_3/Clock_3_YOLO_customized/data.yaml",
    cache=False
)
print(metrics_after_tuned.box.map75)
```

```python
metrics_before_tuned_coco = model_before_tuned.val(
    data=r"/content/drive/MyDrive/object_detection_tuning/coco_val/coco_val_YOLO_original/data.yaml"
)
print (metrics_before_tuned_coco.box.map75)
metrics_after_tuned_coco = model_after_tuned.val(
    data=r"/content/drive/MyDrive/object_detection_tuning/coco_val/coco_val_YOLO_customized/data.yaml",
    cache=False
)
print (metrics_after_tuned_coco.box.map75)
```
